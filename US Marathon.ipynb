##### question from the data

##### difference in speed for the 50km and 50mi male to female

```python
df2.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

##### what group age is the best in the 50m race (20 +races min)(show 15)

```python
df2.query('race_length == "50mi"').groupby('athlete age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending =False).query('count>19').head(15)
```

##### what group age is the worst in the 50m race (20 +races min)(show 20)

```python
df2.query('race_length == "50mi"').groupby('athlete age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending =True).query('count>9').head(20)
```

```python
#Seasons for the data -> Slower in summer than winter?

#spring 3-5
#Summer 6-8
#fall 9-11
#winter 12-2

#split between two decimals
```

```python
df2['race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x:'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'spring' if x>2 else 'Winter')
```

```python
df2.head(25)
```

```python
df2.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending= False)
```
